```python
def w_sum(a, b):
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += (a[i] * b[i])
    return output

weight = [0.1, 0.2, -.1]
def neural_networ(input, weight):
    return w_sum(input, weight)

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

win_or_lose_binary = [1, 1, 0, 1]

true = win_or_lose_binary[0]

input = [toes[0], wlrec[0], nfans[0]]

pred = neural_networ(input, weight)

error = (pred - true) ** 2

delta = pred - true

def ele_mul(num, vector):
    output = [0, 0, 0]
    assert(len(output) == len(vector))
    for i in range(len(vector)):
        output[i] = num * vector[i]
    return output

weight_delta = ele_mul(delta, input)

aplha = 0.01

for i in range(len(weight)):
    weight[i] -= aplha * weight_delta[i]
print(weight)
print(weight_delta)
```

```python
weight = [0.3, 0.2, 0.9]

def neural_networ(input, weight):
    pred = ele_mul(input, weight)
    return pred

wlrec = [0.65, 1.0, 1,0, 0.9]

hurt = [0.1, 0.0, 0.0, 0.1]
win = [1.0, 1.0, 0.0, 1.0]
sad = [0.1, 0.0, 0.1, 0.2]

input = wlrec[0]
true = [hurt[0], win[0], sad[0]]

pred = neural_networ(input, weight)

error = [0, 0, 0]
delta = [0, 0, 0]

for i in range(len(true)):
    error[i] = (pred[i] - true[i]) ** 2
    delta[i] = pred[i] - true[i]

def scalar_ele_mul(number, vector):
    out = [0, 0, 0]
    for i in range(len(vector)):
        out[i] = number * vector[i]
    return out

weight_delta = scalar_ele_mul(input, weight)
alpha = 0.1

for i in range(len(weight)):
    weight[i] -= alpha * weight_delta[i]

print(weight)
print(weight_delta)
```
