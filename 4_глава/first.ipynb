# Градиентный спуск

```python
from math import sqrt


weight = 0.5

alpha = 0.01

def neural_network(weight, input):
    prediction = input * weight
    return prediction

number_of_toes = [8.5]
win_or_lose_binary = [1]

input = number_of_toes[0]
goal_pred = win_or_lose_binary[0]

pred = neural_network(weight, input)

error = (pred - goal_pred) ** 2

delta = pred - goal_pred

weight_delta = delta * input #масщтабирование разности на входе взвешиваемым входом получается мы сразу тут делаем три этапа это масштабирование, обращение знака и остановка

weight -= alpha * weight_delta
```

Итоговая реализация

```python
weight, goal_pred, input = (0.5, 0.8, 0.5)

for i in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = delta * input
    weight -= weight_delta
    print (f'weight: {weight}, error: {error}, delta: {delta}, weight_delta: {weight_delta}, pred {pred}')
```
